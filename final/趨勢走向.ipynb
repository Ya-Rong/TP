```python
import requests
from bs4 import BeautifulSoup
import datetime
from collections import Counter
import matplotlib.pyplot as plt

# 定義目標詞彙
target_word = "加密貨幣"

# 設置爬蟲起始和終止日期
start_date = datetime.date(2023, 1, 1)
end_date = datetime.date(2023, 3, 29)

# 初始化月份和次數的字典
monthly_counts = {}

# 進行網路爬蟲
while start_date <= end_date:
    # 格式化日期
    date_str = start_date.strftime("%Y/%m/%d")
    
    # 構建目標網址
    url = f"https://www.blocktempo.com/{date_str}"
    
    # 發送GET請求
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    
    # 搜尋標題中的主要科技詞彙並統計次數
    titles = soup.find_all("h3", class_="jeg_post_title")
    title_text = ""
    for title in titles:
        title_text += title.text
    words = title_text.split()
    count = Counter(words)
    title_count = count[target_word]
    
    # 將次數加入到對應的月份中
    monthly_counts[start_date.strftime("%Y-%m")] = title_count
    
    # 前進到下個月份
    if start_date.month == 12:
        start_date = datetime.date(start_date.year + 1, 1, 1)
    else:
        start_date = datetime.date(start_date.year, start_date.month + 1, 1)

# 生成趨勢圖
months = list(monthly_counts.keys())
counts = list(monthly_counts.values())

plt.plot(months, counts)
plt.xlabel("Month")
plt.ylabel("Count")
plt.title(f"Monthly Trend of '{target_word}' in News Titles")
plt.xticks(rotation=45)
plt.show()
```
